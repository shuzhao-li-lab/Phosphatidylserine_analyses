````python
import pandas as pd
import os
import shutil
````

````python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
````

````python
seq_df = pd.read_csv("../../../data/input/original_seq/CD8_Tcell_05222018.csv", skiprows=1)
````

````python
# Create a directory
path = "../../../data/input/batch1/raw/raw_renamed/" # You need to copy and paste the whole raw files and create this folder.

#loop through to change name and move to folder

for i in range(seq_df.shape[0]):
    ori_fileName = seq_df.loc[i,"File Name"] # Get the fileName of each instance. This fileName is from the seq_df, so it doesn't have .mzML Suffix.
    new_fileName = seq_df.loc[i,"Sample ID"] # Get the Sample ID of each instance.
    try:
        os.rename(os.path.join(path, ori_fileName) + ".raw", \
              os.path.join(path, new_fileName) + ".raw") 
    except:
        pass
````
