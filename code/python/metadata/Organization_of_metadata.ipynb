```python
# HILIC
HILIC_df['sample_group'] = [("_").join(x.split("_")[0:2]) for x in HILIC_df['Sample ID']]
HILIC_df['technical_group'] = [x.split("_")[2] for x in HILIC_df['Sample ID']]
HILIC_df['sample_name'] = [("_").join(x.split("_")[0:3]) for x in HILIC_df['Sample ID']]
```

## Moving HILIC mzML files
- resources: https://stackoverflow.com/questions/8858008/how-to-move-a-file
- `os.rename("path/to/current/file.foo", "path/to/new/destination/for/file.foo")`

```python
HILIC_df['File Name']
```

```python
os.listdir("../../../data/input/batch1/mzML/")
```

```python
HILIC_df.columns
```

```python
#We are going to change name of File Name to Sample ID.
len(HILIC_df['sample_name'])
```

```python
len(set(HILIC_df['sample_name'])) # To assure we have unique Sample ID
```

```python
# These files will be sent to the same folder named `HILIC_exp` under mzML folder
O_path = "../../../data/input/batch1/mzML/" # define the output path where the new folder will be in.
name_new_fdr = "HILICpos_exp"   # define the new folder name
os.mkdir(O_path + name_new_fdr)
```

```python
#loop through to change name and move to HILIC_exp folder

HILIC_df = HILIC_df.reset_index() #if the index is not reset, you may not be able to loop through
```

```python
for i in range(HILIC_df.shape[0]):
    fileName = HILIC_df.loc[i,"File Name"] # Get the fileName of each instance. This fileName is from the HILIC_df, so it doesn't have .mzML Suffix.
    sample_name = HILIC_df.loc[i,"sample_name"] # Get the Sample ID of each instance.
    os.rename(O_path + fileName + ".mzML", \
              O_path + name_new_fdr + "/" + sample_name + ".mzML") 
```
